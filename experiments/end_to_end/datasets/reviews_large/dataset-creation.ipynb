```python
import random
import pandas as pd
import numpy as np
import datasets
from datasets import load_dataset, Dataset
datasets.get_dataset_config_names("amazon_us_reviews")
```

```python

digital_games = load_dataset("amazon_us_reviews", "Digital_Video_Games_v1_00", split="train")
software = load_dataset("amazon_us_reviews", "Digital_Software_v1_00", split="train")
video_games = load_dataset("amazon_us_reviews", "Video_Games_v1_00", split="train")
```

```python
digital_games = pd.DataFrame(digital_games)
software = pd.DataFrame(software)
video_games = pd.DataFrame(video_games)

combined = pd.concat([digital_games, software, video_games], axis=0)
combined = combined.sample(n=500_000)
combined.to_csv('reviews_3_cats_500000.csv', index=False, sep=";")
combined
```

```python
test_acs = pd.read_csv("reviews_3_cats_500000.csv", delimiter=";")
test_acs
```

```python
# look at reviews small schema
import os
from mlmq.utils import get_project_root
data_root = os.path.join(str(get_project_root()), "experiments", "end_to_end", "datasets")
reviews = pd.read_csv(os.path.join(data_root, "reviews", "reviews.csv.gz"), compression='gzip',
                                      index_col=0)
reviews
```

```python
category_list = test_acs["product_category"].unique()
category_id_map = {'Video Games': 0, 'Digital_Video_Games': 1, 'Digital_Software': 2}
test_acs['category_id'] = test_acs["product_category"].map(category_id_map)
test_acs = test_acs.dropna(subset=["review_body"])
test_acs['category_id']
```

```python
categories = test_acs[['category_id', 'product_category']].reset_index(drop=True)
categories = categories.rename(columns={'category_id': 'id', 'product_category': 'category'}).drop_duplicates().reset_index(drop=True)
products = test_acs[['product_id','product_parent','product_title','category_id']].drop_duplicates().reset_index(drop=True)
ratings = test_acs[['review_id','star_rating','helpful_votes','total_votes']].drop_duplicates().reset_index(drop=True)
reviews = test_acs[['marketplace', 'customer_id', 'review_id', 'product_id', 'vine', 'review_headline', 'review_body', 'review_date']]
reviews['third_party'] = pd.Series(random.choices(['Y', 'N'], k=len(reviews)))
reviews = reviews.reset_index(drop=True)
```

```python
reviews.to_csv(os.path.join(data_root, "reviews_large", "reviews_500000.csv.gz"), index=True, compression='gzip')
ratings.to_csv(os.path.join(data_root, "reviews_large", "ratings_500000.csv"), index=True)
products.to_csv(os.path.join(data_root, "reviews_large", "products_500000.csv"), index=True)
categories.to_csv(os.path.join(data_root, "reviews_large", "categories_500000.csv"), index=True)
```
